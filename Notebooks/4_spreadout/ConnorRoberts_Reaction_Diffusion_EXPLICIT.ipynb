```python
def ftcs(Ui, Vi, nt, Du, Dv, dt, dh, F, k):
    U = Ui.copy()
    V = Vi.copy()
    
    for n in range(nt):
        Un = U.copy()
        Vn = V.copy()
                
        U[1:-1,1:-1] = Un[1:-1,1:-1] + (Du*dt)/(dh**2)*(Un[1:-1,2:]+Un[1:-1,0:-2]+Un[2:,1:-1]+Un[0:-2,1:-1]-4*Un[1:-1,1:-1])\
        - dt*(Un[1:-1,1:-1]*Vn[1:-1,1:-1]**2) + dt*(F*(1-Un[1:-1,1:-1]))
        
        V[1:-1,1:-1] = Vn[1:-1,1:-1] + (Dv*dt)/(dh**2)*(Vn[1:-1,2:]+Vn[1:-1,0:-2]+Vn[2:,1:-1]+Vn[0:-2,1:-1]-4*Vn[1:-1,1:-1])\
        + dt*(Un[1:-1,1:-1]*Vn[1:-1,1:-1]**2) - dt*(Vn[1:-1,1:-1]*(F+k))
        
        #U[1:-2,1:-2] = Un[1:-2,1:-2] + (Du*dt)/(dh**2)*(Un[1:-2,2:-1]+Un[1:-2,0:-3]+Un[2:-1,1:-2]+Un[0:-3,1:-2]-4*Un[1:-2,1:-2])\
        #- dt*(Un[1:-2,1:-2]*Vn[1:-2,1:-2]**2) + dt*(F*(1-Un[1:-2,1:-2]))
        
        #V[1:-2,1:-2] = Vn[1:-2,1:-2] + (Dv*dt)/(dh**2)*(Vn[1:-2,2:-1]+Vn[1:-2,0:-3]+Vn[2:-1,1:-2]+Vn[0:-3,1:-2]-4*Vn[1:-2,1:-2])\
        #+ dt*(Un[1:-2,1:-2]*Vn[1:-2,1:-2]**2) - dt*(Vn[1:-2,1:-2]*(F+k))
  
        # Enforce Neumann BCs
        U[-1,:] = U[-2,:]
        U[:,0] = U[:,1]
        U[0,:] = U[1,:]
        U[:,-1] = U[:,-2]
        
        V[-1,:] = V[-2,:]
        V[:,0] = V[:,1]
        V[0,:] = V[1,:]
        V[:,-1] = V[:,-2]
        
        
    return U
```

```python
n = 192

Du, Dv, F, k = 0.00016, 0.00008, 0.035, 0.065 # Bacteria 1 

dh = 5/(n-1)

T = 8000

dt = .9 * dh**2 / (4*max(Du,Dv))

nt = int(T/dt)
```

```python
Uans = ftcs(Ui, Vi, nt, Du, Dv, dt, dh, F, k)
```
