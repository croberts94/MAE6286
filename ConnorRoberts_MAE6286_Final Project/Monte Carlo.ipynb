## Monte Carlo: EC

```python
import  numpy as np
from random import gauss

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def asset_price(St, sigma, mu, dt):
    St = St * np.exp((mu - 0.5 * sigma**2)*dt + \
                     sigma * gauss(0,1.0) * np.sqrt(dt))
    return St
```

```python
S = 45
K = 40
mu = 0.1
sigma = 0.25
r = 0.1
T = 0.5
N = 100
dt = T/dt
prices = [S]
all_paths = []
St = S
dt

for trial in range (0,5000):
    prices = [S]
    St = S
    for t in range(1,N):
        St = asset_price(St, sigma, mu, dt)
        prices.append(St)
        
    if trial < 1:
        all_paths.append(prices)
    else:
        all_paths = np.vstack((all_paths, prices))
    
    t = range(0,N)
    plt.plot(t,prices)

#plt.plot(t,all_paths[0,:])
#plt.plot(t,all_paths[1,:])
#plt.plot(t,all_paths[2,:])
#plt.plot(t,all_paths[3,:])
#plt.plot(t,all_paths[4,:])
plt.grid(True)
plt.show()
```
